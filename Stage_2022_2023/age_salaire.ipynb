```python
import pandas as pd
import chardet
```

```python
code_commune = pd.read_csv("donnees_communes.csv", sep=";")
salaire = pd.read_csv("cc_base-tous-salaries_2020_COM.csv",sep=";")
chardet.detect(open("age_par_commune.csv", 'rb').read())['encoding']
```

```python
age_par_commune = pd.read_csv("age_par_commune.csv",sep=";",encoding="Windows-1252")
```

```python
code_commune.CODCOM=code_commune.CODCOM.apply(lambda x : str(x).zfill(3))
salaire.CODGEO=salaire.CODGEO.apply(lambda x : str(x).zfill(5))
salaire['code_commune_insee']=salaire.CODGEO
salaire['Salaire']=salaire.SNHM20 #Salaire net horaire moyen en 2020 (euros) ;
code_commune['code_commune_insee']=code_commune.CODDEP+code_commune.CODCOM
code_commune['nom_commune']=code_commune.COM
```

```python
salaire=salaire[["code_commune_insee","Salaire"]]
code_commune=code_commune[["code_commune_insee","PTOT","nom_commune"]]
```
